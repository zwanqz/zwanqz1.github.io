```python
#Khai báo biến dự đoán và biến mục tiêu
X = car_data.drop(['decision'], axis=1)

y = car_data['decision']
```

```python
#kiểm tra lại kiểu dữ liệu của biến
X.dtypes
```

```python
import category_encoders as ce
# mã hóa các biến bằng mã hóa thứ tự
encoder = ce.OrdinalEncoder(cols=['BuyingPrice', 'MaintenanceCost', 'num_Doors', 'num_persons', 'lug_boot', 'safety'])


X = encoder.fit_transform(X)
```

```python
X
```

```python
#Tách dữ liệu - Đào tạo và kiểm tra
#chia X và y thành tập huấn luyện và tập kiểm tra
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# kiểm tra hình dạng của X_train và X_test
X_train.shape, X_test.shape
```

```python
#Entropy tiêu chí phân loại cây quyết định
from sklearn.tree import DecisionTreeClassifier
clf_entropy = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth=3)
#phù hợp với mô hình
clf_entropy.fit(X_train, y_train)
```

```python
y_pred_en = clf_entropy.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Kiểm tra độ chính xác trên dữ liệu thử nghiệm
accuracy_score(y_test,y_pred_en)
```

```python
y_pred_train_en = clf_entropy.predict(X_train)
```

```python
#Kiểm tra độ chính xác của dữ liệu đào tạo
accuracy_score(y_train,y_pred_train_en)
```

```python
#Đánh giá mô hình
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
#Báo cáo phân loại
print(classification_report(y_test, y_pred_en))
```

```python
#Ma trận hỗn loạn
confusion_matrix(y_test, y_pred_en)
```

```python
#Trực quan hóa cây quyết định
plt.figure(figsize = (12,8))
from sklearn import tree

tree.plot_tree(clf_entropy.fit(X_train, y_train))
```

```python
import graphviz
dt = tree.export_graphviz(clf_entropy, out_file = None, feature_names = X_train.columns, class_names = y_train, filled = True, rounded = True, special_characters = True)

graphviz.Source(dt)
```
